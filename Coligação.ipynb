```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coligacao24 = pd.read_csv('consulta_coligacao_2024.csv', sep = ';', encoding='latin-1')
coligacao20 = pd.read_csv('consulta_coligacao_2020.csv', sep = ';', encoding='latin-1')
coligacao16 = pd.read_csv('consulta_coligacao_2016.csv', sep = ';', encoding='latin-1')
coligacao12 = pd.read_csv('consulta_coligacao_2012.csv', sep = ';', encoding='latin-1')
coligacao08 = pd.read_csv('consulta_coligacao_2008.csv', sep = ';', encoding='latin-1')
coligacao04 = pd.read_csv('consulta_coligacao_2004.csv', sep = ';', encoding='latin-1')
coligacao00 = pd.read_csv('consulta_coligacao_2000.csv', sep = ';', encoding='latin-1')
```

```python
total_estados24 = {}
soma_estados24 = {}
estados = ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'ES', 'GO', 'MA', 'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
total24 = 0
soma24 = 0

for x in estados:
  for index, row in coligacao24.iterrows():
    if row['SG_UF'] == x:
      total24 = total24 + 1
      soma24 = soma24 + 1
      total_estados24[x] = total24
      soma_estados24[x] = soma24
      for y in str(row['DS_COMPOSICAO_COLIGACAO']):
        if y == '/':
          soma24 = soma24 + 1
          soma_estados24[x] = soma24

print(total_estados24)
print(soma_estados24)
```

```python
media24 = {}
for x in estados:
  media24[x] = soma_estados24[x] / total_estados24[x]

print(media24)
```
