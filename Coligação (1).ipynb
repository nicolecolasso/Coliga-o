```python
coligacao24 = pd.read_csv('consulta_coligacao_2024.csv', sep = ';', encoding='latin-1')
coligacao20 = pd.read_csv('consulta_coligacao_2020.csv', sep = ';', encoding='latin-1')
coligacao16 = pd.read_csv('consulta_coligacao_2016.csv', sep = ';', encoding='latin-1')
coligacao12 = pd.read_csv('consulta_coligacao_2012.csv', sep = ';', encoding='latin-1')
coligacao08 = pd.read_csv('consulta_coligacao_2008.csv', sep = ';', encoding='latin-1')
coligacao04 = pd.read_csv('consulta_coligacao_2004.csv', sep = ';', encoding='latin-1')
coligacao00 = pd.read_csv('consulta_coligacao_2000.csv', sep = ';', encoding='latin-1')
```

```python
coligacao = pd.concat([coligacao24, coligacao20, coligacao16, coligacao12, coligacao08, coligacao04, coligacao00])
```

```python
estados = ['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'ES', 'GO', 'MA', 'MS', 'MT', 'PA', 'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
```

```python
def media_estado(ano, soma, total, dic1, dic2, dic3):
  for x in estados:
    for index, row in coligacao.iterrows():
      if row['SG_UF'] == x and row['ANO_ELEICAO'] == ano:
        total = total + 1
        soma = soma + 1
        dic1[x] = total
        dic2[x] = soma
        for y in str(row['DS_COMPOSICAO_COLIGACAO']):
          if y == '/':
            soma = soma + 1
            dic2[x] = soma
  for x in estados:
    dic3[x] = dic2[x] / dic1[x]
  return dic3
```

```python
def media_capitais(ano, soma, total, dic1, dic2, dic3):
  for x in capitais:
    for index, row in coligacao.iterrows():
      nm_ue_sem_acento = unicodedata.normalize('NFKD', str(row['NM_UE'])).encode('ASCII', 'ignore').decode('ASCII')
      if nm_ue_sem_acento == x and row['ANO_ELEICAO'] == ano:
        total = total + 1
        soma = soma + 1
        dic1[x] = total
        dic2[x] = soma
        for y in str(row['DS_COMPOSICAO_COLIGACAO']):
          if y == '/':
            soma = soma + 1
            dic2[x] = soma
  for x in capitais:
    dic3[x] = dic2[x] / dic1[x]
  return dic3
```

```python
def media_ano_cap(capital, soma, total, dic1, dic2, dic3):
  for x in anos:
    for index, row in coligacao.iterrows():
      nm_ue_sem_acento = unicodedata.normalize('NFKD', str(row['NM_UE'])).encode('ASCII', 'ignore').decode('ASCII')
      if nm_ue_sem_acento == capital and row['ANO_ELEICAO'] == x:
        total = total + 1
        soma = soma + 1
        dic1[x] = total
        dic2[x] = soma
        for y in str(row['DS_COMPOSICAO_COLIGACAO']):
          if y == '/':
            soma = soma + 1
            dic2[x] = soma
  for x in anos:
    dic3[x] = dic2[x] / dic1[x]
  return dic3
```
